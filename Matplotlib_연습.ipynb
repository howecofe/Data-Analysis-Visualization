```python
# 데이터 읽어 오기. pandas의 read_csv 사용
data = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/data/tips.csv")
data
```

## 1. total_bill 과 tip 의 분포도(scatter) 시각화하기

### tip 금액에 따라 점의 크기 변화 주기, alpha 옵션으로 투명도 지정하기
### 축 label 추가하기, 플롯 title 추가하기


```python
fig = plt.figure(figsize=(8, 6))

x = data['total_bill']
y = data['tip']
sizes = 100 * y

plt.scatter(x, y, c='orange', s=sizes, alpha=0.3)
plt.title('Scatter Plot of Total_bill & Tip')
plt.xlabel('Total_bill')
plt.ylabel('Tip');
```

## 2. 성별에 따른 tip 비교하기 (bar plot)



```python
fig = plt.figure(figsize=(8, 6))

female = data[data['sex']=='Female']['tip'].sum()
male = data[data['sex']=='Male']['tip'].sum()
height = [female, male]
groups = ['Female', 'Male']
y_pos = np.arange(len(groups))

plt.bar(y_pos, height, width=0.5, color='lightgreen')
plt.xticks(y_pos, groups)
plt.title('Bar Plot of Tip according to Sex')
plt.xlabel('Tip');
```

## 3. 남자의 여자의 팁 금액의 분포도(histogram)를 그려 비교하세요

### 2 X 1 의 subplot을 이용하세요.

```python
fig = plt.figure(figsize=(8, 10))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

female = data[data['sex']=='Female']['tip']
male = data[data['sex']=='Male']['tip']

ax1.hist(male, bins=10, color='pink', edgecolor='white')
ax2.hist(female, bins=10, color='skyblue', edgecolor='white')

ax1.set_title('Histogram of Tip according to Sex')
ax1.set_xlabel('Male')
ax1.set_ylabel('Tip')
ax2.set_xlabel('Female')
ax2.set_ylabel('Tip');
```
